```python
#Demonstration of different data types
#Integer
num = 10
print("Integer:", num, type(num))

#Float
pi = 3.14
print("Float:", pi, type(pi))

#String
name = "Tanu"
print("String:", name, type(name))

#Boolean
is_student = True
print("Boolean:", is_student, type(is_student))

#List
my_list = [1, 2, 3, "apple", 4.5]
print("List:", my_list, type(my_list))

#Tuple
my_tuple = (1, 2, 3, "banana")
print("Tuple:", my_tuple, type(my_tuple))

#Dictionary
my_dict = { "name": "Tanu", "age": 18, "course": "Engineering" }
print("Dictionary:", my_dict, type(my_dict))

#Set
my_set = {1, 2, 3, 4, 4, 5}
print("Set:", my_set, type(my_set))
```
